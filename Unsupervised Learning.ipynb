```python
# Import required Libs
import numpy as np
import pandas as pd
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import warnings
```

```python
warnings.filterwarnings('ignore')
```

```python
np.random.seed(0)
data_size = 500
data = {
    'Weight': np.random.randint(1000, 3000, data_size),
    'EngineSize': np.random.uniform(1.0, 4.0, data_size),
    'Horsepower': np.random.randint(50, 300, data_size)
}
```

```python
data
```

```python
df = pd.DataFrame(data)
```

```python
# Preview the dataset, like csv
```

```python
df
```

```python
X = df
```

```python
# Modeling
model = KMeans(n_clusters = 3, random_state=42)
model.fit(X)
```

```python
# Plotting the map
plt.scatter(df['Weight'], df['Horsepower'], c=model.labels_)
plt.xlabel('Weight')
plt.ylabel('Horsepower')
plt.title("Vehicle Cluster")
plt.show()
```
